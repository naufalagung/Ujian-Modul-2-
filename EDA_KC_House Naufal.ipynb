## Lakukan Uji Hipotesis untuk menguji apakah ada perbedaan price antara rumah yang memiliki waterfront dengan yang tidak memiliki waterfront

```python
# STEP 1

df[df['waterfront']==0]['price']

# Memanggil df ga punya waterfront
```

```python
# STEP 2

df[df['waterfront']==0]['price'].var()

# Lihat varians dari df waterfront=o
```

```python
# STEP 3
# Bandingin varians yg ga ada waterfront dengan yg ada waterfont

df[df['waterfront']==0]['price'].var() > df[df['waterfront']==1]['price'].var()
```

```python
# STEP 4
# Kalau beneran lebih besar/kecil, lihat besarnya berapa kali

df[df['waterfront']==0]['price'].var() / df[df['waterfront']==1]['price'].var()

# ini artinya variansnya equal antar kedua variabel karena variansnya dibawah 2 (seinget saya Kak Brigita kemarin bilangnya kalau dibawah 2 artinya equal)
```

```python
# STEP 5
# Uji T-test

result= stats.ttest_ind(df[df['waterfront'] == 0]['price'],
df[df['waterfront']== 1]['price'],
equal_var=True)

print(result)

# karena p-valuenya dibawah 0, maka kita bisa katakan bahwa adanya PERBENDAAN NYATA ketika terdapatnya waterfront terhadap price
```
